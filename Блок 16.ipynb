```python
 def sum_positive_numbers (numbers_list):
     total = 0
     for number in numbers_list:
         if number > 0:
            total += number
     return total
numbers_input = input("Введите числа через пробел: ")
numbers_list = [int(x) for x in numbers_input.split()]

# Вызов функции
result = sum_positive_numbers(numbers_list)
print("Сумма положительных чисел:", result)     
```

```python
def common_keys (dict1, dict2):
        
    result = {}
    
    # Перебираем все ключи первого словаря
    for key in dict1:
        # Если ключ есть во втором словаре И значения совпадают
        if key in dict2 and dict1[key] == dict2[key]:
            result[key] = dict1[key]  # Добавляем в результат
    
    return result

# Проверяем
print(common_keys({'a': 1, 'b': 2}, {'b': 2, 'c': 3}))
```

```python
def sort_numbers(numbers, reverse_order=None ):
    if reverse_order:
        return sorted(numbers, reverse=True)  # по убыванию
    else:
        return sorted(numbers)  # по возрастанию

numbers = [4, 2, 5, 1, 3]         
print(sort_numbers(numbers, False))    
print(sort_numbers(numbers, True))     
```

```python
def compare_elements(list1, list2, comparison_function=None):
    result = []
    
   
    for i in range(min(len(list1), len(list2))):
        elem1 = list1[i]
        elem2 = list2[i]
        
        # Если функция сравнения не указана, используем проверку на равенство
        if comparison_function is None:
            if elem1 == elem2:
                result.append((elem1, elem2))
        else:
            # Используем переданную функцию сравнения
            if comparison_function(elem1, elem2):
                result.append((elem1, elem2))
    
    return result
result1 = (compare_elements([1, 2, 3], [1, 4, 3],comparison_function=lambda x, y: x < y))
print(result1)    
```

```python
def count_unique_chars (*strings):
    # Объединяем все строки в одну
    all_chars = ''.join(strings)
    unique_chars = set(all_chars)
    
    # Возвращаем количество уникальных символов
    return len(unique_chars)
print(count_unique_chars ('hello', 'world'))  
```

```python
def calculate(a, b, operation=None):
    if operation == 'add':
        return a + b
    elif operation == 'subtract':
        return a - b
    elif operation == 'multiply':
        return a * b
    elif operation == 'divide':
        if b == 0:
            return "Ошибка: деление на ноль!"
        return a / b
    elif operation is None:
        return a + b

print(calculate(4, 1))
```

```python
def modify_strings(strings, case='upper'):
    result = []
    for string in strings:
        if case == 'upper':
            result.append(string.upper())  
        elif case == 'lower':
            result.append(string.lower())   
        else:
            result.append(string)           
    return result

print(modify_strings(['Hello', 'World']))
```

```python
def average(*numbers):
    if len(numbers) == 0:
        return 0  # или можно вернуть None, если нет чисел
    
    total = sum(numbers)
    count = len(numbers)
    return total / count
print ( average (10,20))
```

```python
def filter_items (items,condition=None ):
    if condition is None :
        return items
    else :
        result = []
        for item in items:
            if condition(item):
               
                result.append(item)
    return result
print ( filter_items([1, 2, 3, 4, 5], condition=lambda x: x > 3))
```

```python
def find_primes(*arg):
    result=[]
    for i in arg :
        if i%2!=0:
           result.append(i)
    return result
print ( find_primes(1,2,3,4,5,6))
```

```python
def word_count (text,separator=None ):
    s=text.split(separator)
    count={}
    for word in s:
        if word in count:
           count[word]+=1
        else:
           count[word]=1
    return count
print (word_count('red,green,blue,red', separator=','))
```

```python
def filter_numbers (numbers,exclude_multiples_of=2 ):
    result=[]
    for i in numbers :
        if i % exclude_multiples_of!=0:
           result.append(i)
    return result     
print(filter_numbers((1,2,3,4,5,6),3))
```
